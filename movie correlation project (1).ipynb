```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# read csv

df = pd.read_csv(r'C:\Users\Tnluser\Downloads\movies.csv')
```

```python
df.head()
```

```python
# looking for null values in all columns

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# datatypes of our columns

df.dtypes
```

```python
# removing rows having null values

df = df.dropna()
```

```python
# changing datatype of some columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# extracting year from movie release date using regex

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype('int64')
```

```python
# sort df in ascending order of gross revenue generated

df = df.sort_values(by = ['gross'],inplace = False, ascending = False)
```

```python
# code to view the full df

pd.set_option('display.max_rows', None)
# We can use print(df.to_string()) also to print full dataframe
```

```python
# removing duplicates in df

df.drop_duplicates()
```

```python
# scatter plot of budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
```

```python
# reg plot of budget vs gross

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# getting correlation matrix

correlation_matrix = df.corr(method='pearson')
correlation_matrix
```

```python
# heat map of correlation matrix

sns.heatmap(correlation_matrix, annot=True)
```

```python
# conversion of object data types to categories and assigning numerical values to them to get the correlation matrix for all columns

df_num = df
for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name]= df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
```

```python
# correlation matrix for numerized df

cor_matrix = df_num.corr(method = 'pearson')
cor_matrix
```

```python
# heatmap of correlation matrix for numerized df

sns.heatmap(cor_matrix, annot = True)
```

```python
# correlation metric for each column wrt others

cor_pairs = cor_matrix.unstack()
cor_pairs
```

```python
# sorting correlation metric for each column wrt others

sort_pairs = cor_pairs.sort_values()
sort_pairs
```

```python
# filtering the high correlation metrics

high_cor_pairs = sort_pairs[(sort_pairs) > 0.5]
high_cor_pairs
```
